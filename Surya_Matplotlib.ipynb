```python
#website traffic data
#num of users on the wesite

website = [123,645,950,1290,1630,1450,1034,1295,465,205,80]

time = np.arange(7,18)
time
```

```python
style.use('ggplot')

#(x,y) plot

plt.plot(time,website,label="red line")

plt.axis([6.5,17.5,50,2000])

plt.xlabel("time")

plt.ylabel("website customers")

plt.title("website traffic")

plt.legend()
plt.show()
```

```python
style.use('ggplot')

#(x,y) plot

plt.plot(time,website,label="red line",alpha=0.2)


plt.xlabel("time")

#annotate

plt.annotate('Max',ha="center",va="bottom",xytext=(8,1500),xy=(11,1620),arrowprops={'facecolor':'green'})

plt.ylabel("website customers")

plt.title("website traffic")

plt.legend()
plt.show()
```

```python
#multiple plots

#website traffic data multiple days

web_monday=[123,645,950,1290,1630,1450,1039,1295,465,205,80]

web_tuesday=[95,680,889,1145,1620,1343,1119,1265,510,310,110]

web_wednesday=[105,630,700,1006,1520,1124,1239,1380,580,610,230]

time_hrs = np.arange(7,18)

style.use('ggplot')

plt.plot(time_hrs,web_monday,'b',label="Monday",linewidth=2)

plt.plot(time_hrs,web_tuesday,'r',label="Tuesday",linewidth=3)

plt.plot(time_hrs,web_wednesday,'g',label="Wednesday",linewidth=4)

plt.axis([6.5,17.5,50,2000])

plt.xlabel("Time_hrs")

plt.ylabel("Website customers")

plt.title("website traffic")

plt.legend()

plt.show()
```

```python
#subplots
temp=np.arange(75,86)

wind = np.arange(35,46)

time = np.arange(1,12)

humidity = np.arange(54,65)W

precipitation = np.arange(22,33)

#subplot for 1,2,1 and 1,2,2

plt.figure(figsize=(8,4))
plt.subplots_adjust(hspace=0.25)

plt.subplot(1,2,1)
plt.title('Temperature')
plt.plot(time,temp,color='g',ls='-',lw=1)

plt.subplot(1,2,2)

plt.title('Humidity')

plt.plot(time,humidity,color='r',lw=2)
```

```python
#import boston data set

from sklearn.datasets import load_boston

df = load_boston()

import matplotlib.pyplot as plt

from matplotlib import style

%matplotlib inline

print(df.DESCR)
```

```python
#x_axis 

x_axis = df.data

#y_axis

y_axis = df.target

#plot histogram

style.use('ggplot')

plt.figure(figsize=(7,7))

plt.hist(y_axis,bins=50)

plt.xlabel("Price of 1000s USD")

plt.ylabel("Number of houses")

plt.show()
```

```python
#plot scatter plot

style.use('ggplot')

plt.figure(figsize=(6,6))

plt.scatter(df.data[:,5],df.target,color='b')

plt.ylabel("Price of 1000s USD")

plt.xlabel("Number of houses")

plt.show()
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0.1,4,0.5)
y=np.exp(x)

plt.errorbar(x,y,xerr=0.4,yerr=0.5)

plt.title("Error Bar")

plt.show()
```

```python
#multiple plots in error bars

x = np.arange(10)

y = np.sin(x/20 * np.pi)

yerr = np.linspace(0.05,0.2,10)

plt.errorbar(x,y+7,yerr=yerr,label="Line1",color='r')

plt.errorbar(x,y+5,yerr=yerr,uplims=True,label="Line2",color='b')

plt.errorbar(x,y+3,yerr=yerr,uplims=True,lolims=True,label="Line3",color='m')

upperlimits = [True,False]*5
lowerlimits = [False,True]*5

plt.errorbar(x,y,yerr=yerr,uplims=upperlimits,lolims=lowerlimits,label="Line4",color='y')
plt.legend()
plt.show()
```
