```python
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
listings=pd.read_csv("listings.csv")
```

```python
listings
```

```python
sb.countplot(x='neighbourhood_group', data=listings)
plt.show()
```

```python
sb.barplot(x='neighbourhood_group',y='price',data=listings,ci=False)
plt.show()
```

```python
plt.hist(listings['price'],bins=np.arange(0,1100,40))
plt.xlabel('price[in us dollars]')
plt.show()
```

```python
plt.scatter(x=listings['price'],y=listings['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.show
```

```python
plt.scatter(x=listings['price'],y=listings['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.xlim(0, 1100)
plt.show
```

```python
plt.scatter(x=listings['price'],y=listings['number_of_reviews'],s=5)
plt.xlabel('price')
plt.ylabel('number of reviews')
plt.xlim(0, 1100)
plt.show
```
