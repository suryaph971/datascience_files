```python
# data operations

# example :  movie rating


import pandas as pd


df  = pd.DataFrame({'movie1':[5,4,3,2,1],
                    'movie2':[3,4,2,1,5]}
                    ,index=['Tom','Red','Jack','Nein','reme'])

df


#applymapmethod


def movierating(rating):
  if rating == 5:
    return 'A'
  elif rating == 4:
    return 'B'
  elif rating == 3:
    return 'C'
  elif rating == 2:
    return 'D'
  else:
    return 'E'


df.applymap(movierating)
```

```python
#data standardization

import pandas as pd

df = pd.DataFrame({'Test1':[30,20,45,55,67],
                   'Test2':[34,67,56,53,11]},
                   index=['Surya','Anees','Mohith','Indra','Riyaz'])

def standardize_test(Test):
  return (Test - Test.mean()) / Test.std()

standardize_test(df['Test1'])


def standardize_test1(datafrm):
  return datafrm.apply(standardize_test)  #apply function to emtire data set

standardize_test1(df)
```

```python
# data operations - part2

import pandas as pd


df = pd.DataFrame({'Student':['Tom','Ram','Tuwen','Skiey','Jacob'],
                   'ID':[20,30,40,45,67]})

df1 = pd.DataFrame({'Student':['Tom','Jacob','Ram'],
                    'ID':[20,24,45]})

```

```python
#merging the data sets
pd.merge(df,df1)
```

```python
#merging with the key
pd.merge(df,df1,on='Student')
```

```python
# merge as  left join

pd.merge(df,df1,on='ID',how='left').fillna('X')
```

```python
#concatenate data frames

pd.concat([df,df1],ignore_index='True')
```

```python
#pandas Sql operation

import pandas as pd

import sqlite3 as sql

#create sql table

createtable = """
Create Table Student_score(Id integer, Name Varchar(20), Math REAL, Science REAL);"""

# executing the sql statement

st = sql.connect(':memory:')
st.execute(createtable)
st.commit()

#sql query
sql_query = st.execute('select * from Student_score')

result = sql_query.fetchall()

result

```

```python
insertsql=[(10,'Surya',98,97),(11,'Nikhil',94,92),(12,'Ramesh',91,90),(13,'Kiran',93,91)]

st1 = 'Insert into Student_Score values(?,?,?,?)'
st.executemany(st1,insertsql)
st.commit()

sql_query = st.execute('select * from Student_score')

result = sql_query.fetchall()

result
```

```python
#put the records in data frame

df = pd.DataFrame(result,columns= zip(*sql_query.description)[0])
df
```
