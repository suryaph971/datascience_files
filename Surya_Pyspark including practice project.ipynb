```python
from pyspark import SparkConf
from pyspark.context import SparkContext as sc
```

```python
#create an RDD with text file method
s = sc.getOrCreate(SparkConf())
data_file = s.textFile("random.txt")
```

```python
#importing operator lib for word  count operation
from operator import add
```

```python
#use filter to return RDD for words length greater than 0
data_filter = data_file.filter(lambda x : len(x)>0)
```

```python
#verify the word count

data_filter.count()
```

```python
#use flatmap to split each line into words
data_filter = data_filter.flatMap(lambda x :x.split())
```

```python
#map each key with value 1 using map function

data_filter = data_filter.map(lambda x : (x,1))
```

```python
#use reduceByKey to reduce above mapped keys
#return key-value pairs by adding values for above keypairs

data_filter = data_filter.reduceByKey(add)
```

```python
#view the first element

data_filter.first()
```

```python
#view first 3 key value pairs

data_filter.take(3)
```
